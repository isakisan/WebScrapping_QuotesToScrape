```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
import pandas as pd
```

```python
tag_name="life"
page=1
url = f"https://quotes.toscrape.com/tag/{tag_name}/page/{page}/"
```

```python
r=requests.get(url)
soup=bs(r.content,"html.parser")
```

```python
print(r)
```

```python
print(soup)
```

```python
div1 = soup.find_all('div',{"class":"col-md-8"})
```

```python
small1 = soup.find_all('small', {"class":"author"})
```

```python
print(div1)
```

```python
print(small1)
```

```python
# Extract the HTML content from the ResultSet
html_content = ''.join(map(str, div1))

# Create a BeautifulSoup object
soup = bs(html_content, 'html.parser')
```

```python
soup
```

```python


# Find all 'div' elements with class 'quote'
quote_elements = soup.find_all('div', class_='quote')

# Iterate through each quote element
for i, quote_element in enumerate(quote_elements, 1):  # Assuming you want 1-based index
    # Extract the quote text
    quote_text = quote_element.find('span', class_='text').text.strip()

    # Extract the author name
    author_name = quote_element.find('small', class_='author').text.strip()

    # Extract the tags
    tags = [tag.text for tag in quote_element.find_all('a', class_='tag')]

    # Print the information
    print(f"[{i}] Author: {author_name}")
    print(f"    Quote: {quote_text}")
    print(f"    Tags: {', '.join(tags)}\n")
```

```python
#storing the scrapped data in a list
Author_Names=[]
Quotes=[]
Tags=[]
for i, quote_element in enumerate(quote_elements, 1):# Assuming you want 1-based index
    # Extract the quote text
    quote_text = quote_element.find('span', class_='text').text.strip()

    # Extract the author name
    author_name = quote_element.find('small', class_='author').text.strip()

    # Extract the tags
    tags = [tag.text for tag in quote_element.find_all('a', class_='tag')]
    
    Author_Names.append(author_name)
    Quotes.append(quote_text)
    Tags.append(tags)
```

```python
print("Authors:", Author_Names)
print("Quotes:", Quotes)
print("Tags:", Tags)
```

```python
data={'Authors':Author_Names, 'Quotes':Quotes, "Tags":Tags}
```

```python
df=pd.DataFrame(data)
```
